# Import Libraries

```python
import pandas as pd
import numpy as np
print('upload library')
```

# Load the Blinkit dataset

```python
df=pd.read_csv('Blinkit_Grocery.csv')
print('upload csv')
```

# Initial check for missing values per column

```python
df.isnull().sum()
```

# Clean and Verify Item Weight

```python
df['Item_Weight']= df['Item_Weight'].fillna(df['Item_Weight'].mean())
df.isnull().sum()
```

# Clean and Verify Outlet Size

```python
df['Outlet_Size']=df['Outlet_Size'].fillna('medium')
df.isnull().sum()
```

# Standardize and Verify Fat Content Categories

```python

# Inspect unique categories and their counts before cleaning
print("Before Standardization:")
print(df['Item_Fat_Content'].value_counts())

# Standardize inconsistent names to 'Low Fat' and 'Regular'
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({'LF': 'Low Fat', 'low fat': 'Low Fat', 'reg': 'Regular'})

# Inspect unique categories to verify the clean output
print("\nAfter Standardization:")
df['Item_Fat_Content'].value_counts()
```

# Engineer and Verify Estimated Profit Feature

```python
# Create a new feature calculating an estimated 15% profit margin on sales
df['Estimated_Profit'] = df['Item_Outlet_Sales'] * 0.15

# View the original sales and new profit column together to verify the math
df[['Item_Outlet_Sales', 'Estimated_Profit']].head()
```

# Export Cleaned Dataset for SQL

```python
# Save the fully transformed and verified dataset to a new CSV file
df.to_csv('blinkit_cleaned_final.csv', index=False)
print("Success: 'blinkit_cleaned_final.csv' has been created and saved!")
```
